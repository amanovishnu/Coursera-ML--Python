```python
# Sine Graph
import numpy as  np
import matplotlib.pyplot as plt
t = np.arange(0.0,3.0,0.01)
s = 1 +np.sin(2*np.pi*t)
plt.plot(t,s,'r')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage in (mA)')
plt.title('Voltage vs Time Sine Graph')
plt.grid()
plt.show()
```

```python
# Cosine Graph
t = np.arange(0.0,3.0,0.01)
s = 1+np.cos(2*np.pi*t)
plt.plot(t,s,'g')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (mA)')
plt.title('Voltage vs Time Cosine Graph')
plt.grid()
plt.show()
```

```python
# Tan Graph
t = np.arange(0.0,3.0,0.01)
s = 1+np.tan(2*np.pi*t)
plt.plot(t,s,'black')
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (mA)')
plt.title('Voltage vs Time Tan Graph')
plt.grid()
plt.show()
```
