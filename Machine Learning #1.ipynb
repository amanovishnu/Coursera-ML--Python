```python
import sys, os, matplotlib, matplotlib.pyplot as plt, numpy as np
print('System Version : ',sys.version)
print('System Version Info : ',sys.version_info)
print('System Platform : ',sys.platform)
print('Numpy Version : ', np.__version__)
print('MatplotLib Version : ', matplotlib.__version__)
```

```python
# SIne Graph Using np.sin()
x1 = np.arange(0.0,5.0,0.001)
# print(x1) 
y1 = np.sin(2*np.pi*x1)
# print(y1)
plt.plot(x1,y1,color='red')
plt.xlabel('Time in (msec)')
plt.ylabel('Amplitude in (mA)')
plt.title('Sine graph Using Numpy Library')
plt.grid()
plt.show()
```

```python
# Cos Graph Using np.cos()
x2 = np.arange(0.0,5.0,0.001)
# print(x2)
y2 = np.cos(2*np.pi*x2)
# print(y2)
plt.plot(x2,y2,color='blue')
plt.xlabel('Time in (msec)')
plt.ylabel('Amplitude in (mA)')
plt.title('Cos Graph using Numpy Library')
plt.grid()
plt.show()
```

```python
# Tan Graph Using np.tan()
x3 = np.arange(0.0,5.0,0.001)
# print(x3)
y3 = np.tan(2*np.pi*x3)
# print(y3)
plt.plot(x3,y3,color='black')
plt.xlabel('Time in (msec)')
plt.ylabel('Amplitude in (mA)')
plt.title('Tan Graph using Numpy Library')
plt.grid()
plt.show()
```
