```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)
plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')
plt.title('Subplot 1')
plt.xlabel('x1')
plt.ylabel('y1')
plt.subplot(2,1,2)
plt.plot(x2,y2,'.-')
plt.title('Subplot 2')
plt.xlabel('x2')
plt.ylabel('y2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as  plt
t1 = np.arange(0.0,5.0,0.01)
s1 = 1+np.sin(2*np.pi*t1)
t2 = np.arange(0.0,2.0,0.02)
s2 = 1+np.cos(2*np.pi*t2)
plt.subplot(2,1,1)
plt.plot(t1,s1,'r')
plt.title('Sin vs Time Graph')
plt.xlabel('Voltage (mA)')
plt.ylabel('time (ms)')
plt.grid()
plt.show()
plt.subplot(2,1,2)
plt.plot(t2,s2,'black')
plt.title('Cos vs Time Graph')
plt.xlabel('Voltage (mA)')
plt.ylabel('time (ms)')
plt.grid()
plt.show()
```
