```python
import os, sys, numpy as np, scipy, matplotlib, matplotlib.pyplot as plt, sklearn
print('System Version : ',sys.version)
print("System Version Info : ",sys.version_info)
print('Numpy Version : ',np.__version__)
print('Scipy Version : ',scipy.__version__)
print('Matplotlib Version : ',matplotlib.__version__)
print('Scikit Learn Version : ',sklearn.__version__)
```

```python
# Plotting Multiple Graphs on Single plane / pane
# time = np.linspace(0.0,5.0, num=50)
time= np.arange(0.0,5.0,0.01)
# print(time)
sin = np.sin(2*np.pi*time)
cos = np.cos(2*np.pi*time)
tan = np.tan(2*np.pi*time)
user = np.linspace(0.0,5.0,num=500)
# print(sin)
plt.subplot(2,2,1)
plt.plot(time,sin,'-',color="red", label="Sin")
plt.xlabel('Time in msec')
plt.ylabel('Amplitude in mA')
plt.title('Sin Graph')
plt.grid()
plt.legend()
plt.subplot(2,2,2)
plt.plot(time,cos,'-',color='black', label="Cos")
plt.xlabel('Time in msec')
plt.ylabel('Amplitude in mA')
plt.title('Cos Graph')
plt.grid()
plt.legend()
plt.subplot(2,2,3)
plt.plot(time,tan, '-', color='violet' ,label="Tan")
plt.xlabel('Time in msec')
plt.ylabel('Amplitude in mA')
plt.title('Tan Graph')
plt.grid()
plt.legend()
plt.subplot(2,2,4)
plt.plot(time,user, '-',color="orange", label ='User Function')
plt.xlabel('Time in msec')
plt.ylabel('Amplitude in mA')
plt.title('User Function Graph')
plt.grid()
plt.legend()
plt.show()
```
